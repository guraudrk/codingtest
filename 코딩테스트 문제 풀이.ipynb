```python
#백준 17298번(골드4)-2024/07/06- 정답

#이 과정을 완전히 이해하면 솔직히 코테 걱정은 없다.

#백준 17298번(골드4)-2024/07/06

#문제의 핵심은 스택에 인덱스를 저장해야 한다는 것이다. 그래야 편하다.


#1.배열 A 정의/수열크기 정의/정답배열정의(answer)

N = int(input())
A =list(map(int,input().split()))

answer =[-1]*N #정답 출력하는 배열
stack = [] #스택을 정의하는 배열

for i in range(N):
    while stack and A[stack[-1]]<A[i]: 
        #처음에는 stack이 비어 있으니까, 당연히 0을 append 하겠지?
        answer[stack.pop()] = A[i] #stack.pop을 코드 안에 추가함으로서 스택을 자동으로 비운다.
    stack.append(i) #인덱스를 추가한다. 
    
    
print(*answer)
```
