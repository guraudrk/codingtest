```python
#백준 1260번(실버2)-2024/07/16

#아이디어
#1.dfs와 bfs의 성질을 잘 이해하면 쉽게 풀 수 있는 문제이다.
#2.dfs는 스택(후입선출),bfs는 큐(선입선출)를 활용해야 한다.
#3.dfs를 호출할 때에만 결과 배열에 값을 추가한다.
#4.마지막에, 배열의 값들을 한번에 출력할 때 *을 쓴다.
#5.재귀함수 호출 제한-sys.setrecursionlimit을 쓴다.
import sys
sys.setrecursionlimit(10**6)

from collections import deque #파이썬에서 큐를 사용하기 위한 코드.
#숫자들을 입력한다.
a,b,c = map(int,input().split())

#2차원 배열은 이런 식으로 정의를 한다.
A=[[]for i in range(a+1)]

#bfs,dfs의 결과를 저장할 배열들을 정의한다.

depth=[] #dfs(후입선출)
breadth=[] #bfs(선입선출)
stack=[] #dfs에서 사용할 스택.
q=deque() #bfs에서 사용할 큐.
iscalleddfs=[0]*(a+1)#각 숫자가 호출이 되었는가? 호출시 값을 1로 변환
iscalledbfs=[0]*(a+1)
for i in range(b):
    n,m = map(int,input().split())
    A[n].append(m)
    A[m].append(n)

#print(A) #배열에 입력값이 잘 들어가 있나? 확인

#배열을 정렬한다.
for i in range(a+1):
    A[i]=sorted(A[i])

#print(A)


#dfs를 먼저 정의한다. dfs는 후입선출/스택을 쓴다.
#중요한 것은 처음 다루는 숫자에 한해서 stack에 넣거나 새로운 dfs 함수를 호출한다거나 하는 것이다.
def dfs(n):
    if(len(iscalleddfs)==a):
        return #모두 집어 넣었다면 return으로 호출 끝냄.
    if(iscalleddfs[n]==0): 
        stack.append(n)
        depth.append(n)
        iscalleddfs[n]=1 
    for i in A[n]:
        if(iscalleddfs[i]==0):
            stack.append(i)
            #depth.append(i)
            iscalleddfs[n]=1 
            k=stack.pop()
            dfs(k)


#그 다음에 bfs를 정의한다. bfs는 선입선출/큐를 쓴다.
def bfs(n):
    if(len(iscalledbfs)==a):
        return #모두 집어넣었다면 return으로 호출을 끝냄.
    if(iscalledbfs[n]==0):
        breadth.append(n)
        q.append(n)
        iscalledbfs[n]=1
    for i in A[n]:
        if(iscalledbfs[i]==0):
            q.append(i)
            #breadth.append(i)
            iscalledbfs[n]=1
    q.popleft() #큐의 가장 먼저 들어가 있는 놈을 꺼낸 뒤 지운다.
    if len(q)==0:
        return
    bfs(q[0])#dfs와 재귀함수를 호출하는 타이밍이 다르다.    
    
#값을 출력한다.
dfs(c)
bfs(c)

print(*depth)
print(*breadth)
```

```python
#백준13023번(골드5)-2024/07/20


#아이디어
#이 문제는 dfs로 푸는 문제이다.
#1.for i in range(n)을 정의한다. visited 배열도 정의한다.
#2.A[i]의 숫자들을 탐방한다. visited가 방문하지 않은 곳으로 간다.
#3.방문시 count++하고 A[k]로 이동한다.
#4.2,3을 반복한다. 
#5.dfs의 count가 5가 될 때 1을 출력한다.
#6.핵심은 dfs(i)가 아니라 dfs(i,count=0)와 같이 정의해서, 깊이 변수를 다루는 것이다. 깊이 변수를 전역 변수로 정의하면 골치아프다.

#오류 및 해결
#1.arrived=1이 되었는데 막상 print을 할 때는 0이 나왔다.arrived를 정수형으로 선언하지 말고 True/False로 선언하면 오류를 막을 수 있다. 또한 변수를 global로 선언해야 한다.
#2.dfs를 선언하고 바로 다음 줄에도 visited[i]=1을 해야 한다. 이걸 까먹으면 섭하지.
#3.dfs에서, 다른 dfs를 호출하지 못했다면 그냥 무위로 돌아간 것이므로, visited를 false로 한다.
#4.백트래킹을 활용해야 한다. 안그러면 다른 경우의 수에서 탐색을 원활히 할 수 없다.

import sys
sys.setrecursionlimit(10000) #재귀 문제를 풀 때 이 변수를 정의하는 것이 절대적이다.
input=sys.stdin.readline


N,M = map(int,input().split())

A=[[] for _ in range(N+1)]

for i in range(M):
    a,b = map(int,input().split())
    A[a].append(b)
    A[b].append(a)


#print(A) #A가 잘 들어가 있나?

arrived=False #깊이가 5가 된 적이 있나?
visited=[0]*(N+1)
def dfs(i,count):
    visited[i]=1
    global arrived
    #print(count)
    if count==5:
        arrived=True #여기서 arrived를 1로 바꿨는데 밑의 arrived는 0이 나온다. 해결:arrived를 True/False로 바꾼다.
        return
        #print(arrived)
    for k in A[i]:
        if(visited[k]==0):
            #visited[k]=1
            dfs(k,count+1)
    visited[i]=0 #다른 경로에서도 재탐색이 가능하도록 하는 코드. 이를 백트레킹이라고 한다.
for i in range(N):
    #방문했는지 안했는지 탐색하는 배열.
   
    dfs(i,1)
    if arrived:
        break
    
if arrived:
    print(1)
else:
    print(0)
```

```python
from collections import deque

Sender = [0,0,1,1,2,2]
Receiver = [1,2,0,2,0,1] #물을 이동시키는 6가지 경우의 수.

now=list(map(int,input().split()))
visited=[[False for j in range(201)]for i in range(201)] #문제에 나와 있는 대로 200이 한계. 이는 물통 A,B의 무게들이 이미 나온 것인지 체크하는 배열.
answer = [False]*201


def BFS():
    queue = deque()
    queue.append((0,0))
    visited[0][0]=True #물통 A,B모두 0인 경우를 true로 마킹.
    answer[now[2]]=True #초기의 C의 값도 true로 마킹.
    
    
    while queue:
        now_Node = queue.popleft()
        A = now_Node[0]
        B = now_Node[1]
        C = now[2]-A-B #c는 전체 물의 양에서 a와 b를 뺀 것
        for k in range(6):
            #물통을 이동시키는 6가지의 수를 구현한다.
            next = [A,B,C]
            next[Receiver[k]]+=next[Sender[k]] #sender가 Receiver에게 물을 다 준다.
            next[Sender[k]] = 0 #물통을 비웠으므로 0으로 친다.
            
            if next[Receiver[k]]>now[Receiver[k]]:
                #물이 넘칠 때를 정의한다.
                #이 때는 초과하는 만큼 다시 이전 물통에 넣어준다.
                next[Sender[k]] = next[Receiver[k]]-now[Receiver[k]]
                next[Receiver[k]] = now[Receiver[k]] #대상 물통 최대로 채우기
            if not visited[next[0]][next[1]]:
                visited[next[0]][next[1]] = True
                queue.append((next[0],next[1]))
                if next[0] == 0: #a의 물통이 가득할 경우에 물통 c의 값을 true로 한다.
                    answer[next[2]] = True


BFS()

for i in range(len(answer)):
    if answer[i]:
        print(i,end=' ')
```

```python
#백준 1516번(골드3)-2024/08/14


#아이디어
#1.위상 정렬을 통해 문제를 푼다.
#2.문제는 어떻게 입력을 하느냐는 것이다.
#3.배열 b의 '이 건물을 선행 건물로 갖는 건물들의 리스트'라는 의미를 잘 보자.
#4.중요한 것은 정답 리스트에 자기 건물을 짓는 시간을 더하는 것이다.
#5.정답 리스트를 갱신할 때 주의해서 갱신해야 한다.
#6.이 문제에서 진입


from collections import deque
N = int(input()) #건물 종류 수
queue = deque() #큐(선입선출이므로 위상정렬에 많이 쓰인다)를 정의하자.

time=[0]*(N+1) #건물 짓는 시간 리스트 
answer_time = [0]*(N+1) #정답 리스트
A=[0]*(N+1) #진입 차수 리스트
B=[[] for i in range(N+1)] #이 건물을 선행 건물로 갖는 건물들의 리스트


for i in range(1,N+1):
    numbers = list(map(int,input().split())) #숫자들을 입력받는다.
    time[i]=numbers.pop(0)
    while numbers:
        number=numbers.pop(0) #숫자를 꺼낸다.
        if number==-1:
            break #-1을 입력하면 끝나는 것이다.
        B[number].append(i) 
        A[i]=A[i]+1 #진입 차수에 추가된다.




#우선 진입 차수가 0인 것들 먼저 출력한다.
for i in range(1,N+1):
    if(A[i]==0):
        queue.append(i) #진입 차수가 0인 건물들 먼저 queue에 append
    
#정답 리스트를 갱신할 때 주의해서 갱신해야 한다.        
while queue:
    now = queue.popleft() #먼저,값을 꺼낸다. 
    #print(time[now]+answer_time[now]) #시간을 출력한다.
    for next in B[now]:
        answer_time[next] = max(time[now]+answer_time[now],answer_time[next]) #이렇게 더해야 누적의 개념이 되는 것
        #print("answer_time:",answer_time)
        A[next]=A[next]-1 #진입 차수에 1을 뺀다.
        if A[next]==0:
            queue.append(next)
            
    
#print("answer_time:",answer_time)
for i in range(1,N+1):
    print(answer_time[i]+time[i])
    

```

```python
#백준 11404번(골드4)-2024/08/15

#아이디어 
#1.이 문제는 플로이드 워셜을 활용하여 문제를 푼다.
#2.최대값을 정의할 때 sys.maxsize를 정의하면서 문제를 풀자.
#3.점화식: D[S][E] =min(D[S][E],D[S][K]+D[K][E]) (3중 for문 사용)-문제를 풀 때는 다르게 구현하면 된다.
#4.i에서 j로 갈 수 없을 때는 0을 출력한다는 부분을 조심해서 문제를 풀자.(2중 for문을 다시 활용해서 문제를 풀 때, 값이 sys.maxsize와 같으면 0을 대신 출력)
#5.그냥 한줄 띄울 때는 print()을 쓴다.
#6.그래프에 비용 입력하기 전에 무작정 입력하지 말고 if문을 통해 대소관계를 먼저 따지기.


import sys
input = sys.stdin.readline
N=int(input()) #도시 개수
M=int(input()) #노선 개수
A=[[sys.maxsize for j in range(N+1)]for i in range(N+1)] #문제에 쓸 리스트를 정의한다.

#그래프를 정의하기 전에, 대각선을 모두 0으로 초기화한다. 자기 자신은 갈 수 없기 때문이다.
for i in range(1,N+1):
    A[i][i]=0 #대각선 부분을 0으로 바꾼다.


for i in range(M):
    a,b,c = map(int,input().split()) #시작도시/도착도시/비용을 정의한다.
    if(A[a][b]>c):
        A[a][b]=c #그래프에 비용을 입력한다.
    
#print(A)   

#플로이드-워셜은 3중 for문을 통해 문제를 해결한다.


for K in range(1,N+1):
    for S in range(1,N+1):
        for E in range(1,N+1):
            if(A[S][E]>A[S][K]+A[K][E]):
                A[S][E]=A[S][K]+A[K][E]
            #A[S][E]=min(A[S][E],A[S][K]+A[K][E]) #점화식을 통해 값을 경신한다.
            
            
#값을 출력한다. 이 때, 값이 sys.maxsize와 같다면 경로가 없다는 의미이므로 0을 대신 출력한다.
for S in range(1,N+1):
    for E in range(1,N+1):
            if(A[S][E]==sys.maxsize):
                print(0,end=' ')
            else:
                print(A[S][E],end=' ')
    print() #그냥 한줄 띄고 싶을 때는 print()을 쓴다.
```

```python
#백준1197(골드4)-2024/08/21

#아이디어
#1.우선순위 큐를 통해 가중치가 작은 순으로 정렬한다.
#2.union,find을 통해 연결을 하고, 사이클이 발생하지 않으면 에지를 추가한다.
#3.에지를 더한 횟수가 '노드 개수-1'이 될 때까지 반복하고, 반복이 끝나면 에지의 가중치를 모두 더한 값을 출력한다.


from queue import PriorityQueue
import sys

input = sys.stdin.readline

pq = PriorityQueue() #시작/도착 노드, 가중치들을 저장할 우선순위 큐


#노드 개수, 에지 개수를 구한다. 
N,M = map(int,input().split())

parent = [0] * (N+1)


for i in range(N):
    parent[i] = i #parent 배열의 값들을 초기화한다.
    
    

for i in range(M):
    s,w,e = map(int,input().split()) #값들을 입력한다.  
    pq.put((e,s,w))
    


def find(a):
    if a == parent[a]:
        return a
    else:
        parent[a] = find(parent[a])
        return parent[a]
        
def union(a,b):
    a=find(a)
    b=find(b)
    if a!=b:
        parent[b]=a
        

#'에지를 더한 횟수'를 정의하는 변수를 정의하는 것이 좋다.
edge = 0
total = 0 #이거는 값을 합친거.
while edge<N-1:
    e,s,w = pq.get() #가중치가 작은 순대로 값을 받는다.
    if(find(s)!=find(w)):
        union(s,w)
        edge = edge + 1
        total= total + e
        
print(total)
```

```python
#백준 11437번(골드3)-2024/08/23
#1.공통 조상을 구하는 알고리즘을 잘 활용해야 한다. 트리 문제에 필수적인 BFS/DFS+인접리스트는 필수.
#2.BFS의 코드가 매우 이해하기 난해한데, 이를 잘 이해해야 한다.
#3.depth을 맞추는 코드와 parent을 맞추는 코드는 은근히 쉬우니, 잘 거시기하자.
#4.BFS에서는 현재 깊이에서 처리해야 할 총 노드의 개수/지금까지 해치운 노드의 개수를 각각 변수를 정의해서 처리해야 한다.
import sys
from collections import deque
input = sys.stdin.readline
print = sys.stdout.write
N = int(input())
tree = [[] for _ in range(N+1)]

#인접 리스트에 트리 데이터 저장
for _ in range(0,N-1):
    s,e = map(int,input().split())
    tree[s].append(e)
    tree[e].append(s)


#깊이*부모 노드*방문 배열을 정의한다.
depth = [0]*(N+1)
parent = [0]*(N+1)
visited=[False]*(N+1)


def BFS(node):
    queue = deque() #큐 정의
    queue.append(node) #큐에 노드를 추가한다.
    visited[node] = True
    level=1
    now_size=1
    count=0
    
    #큐 안에 있는 것들을 하나씩 꺼낸다.
    while queue:
        now_node = queue.popleft()
        for next in tree[now_node]:
            if not visited[next]:
                visited[next]=True
                queue.append(next)
                parent[next]=now_node #부모 노드 배열에 값을 저장한다.
                depth[next]=level #노드의 depth을 저장한다.
        count+=1 #현재 레벨에서 몇 개의 노드를 처리했는지
        if count == now_size: #현재 레벨의 모든 노드를 처리한 경우
            count=0 #count를 초기화한다.
            now_size = len(queue) #다음 레벨에서 처리해야 할 노드 수로 업데이트 한다.
            level+=1 
                
BFS(1) #문제에서 루트는 1번이라고 했기에, 이렇게 지정한다.



def executeLCA(a,b):
    if depth[a]<depth[b]:
        temp = a
        a = b
        b = temp
    #큰 쪽이 앞에 가서 계산을 진행하도록 기준을 맞추기 위한 코드. 둘이 바꾸는 코드는 잘 알지?
    
    while depth[a]!=depth[b]:
        #depth 맞추기
        a= parent[a]
    while a!=b:
        a=parent[a]
        b=parent[b]
        #공통 조상 맞추기
    return a #공통 조상을 리턴한다.
    
M=int(input())


for _ in range(M):
    a,b = map(int,input().split())
    print(str(executeLCA(a,b)))
    print("\n") #띄어쓰기
    

                
                
```
