```python
#백준 17298번(골드4)-2024/07/06- 정답

#이 과정을 완전히 이해하면 솔직히 코테 걱정은 없다.

#백준 17298번(골드4)-2024/07/06

#문제의 핵심은 스택에 인덱스를 저장해야 한다는 것이다. 그래야 편하다.


#1.배열 A 정의/수열크기 정의/정답배열정의(answer)

N = int(input())
A =list(map(int,input().split()))

answer =[-1]*N #정답 출력하는 배열
stack = [] #스택을 정의하는 배열

for i in range(N):
    while stack and A[stack[-1]]<A[i]: 
        #처음에는 stack이 비어 있으니까, 당연히 0을 append 하겠지?
        answer[stack.pop()] = A[i] #stack.pop을 코드 안에 추가함으로서 스택을 자동으로 비운다.
    stack.append(i) #인덱스를 추가한다. 
    
    
print(*answer)
```

```python
#백준 1260번(실버2)-2024/07/16

#아이디어
#1.dfs와 bfs의 성질을 잘 이해하면 쉽게 풀 수 있는 문제이다.
#2.dfs는 스택(후입선출),bfs는 큐(선입선출)를 활용해야 한다.
#3.dfs를 호출할 때에만 결과 배열에 값을 추가한다.
#4.마지막에, 배열의 값들을 한번에 출력할 때 *을 쓴다.
#5.재귀함수 호출 제한-sys.setrecursionlimit을 쓴다.
import sys
sys.setrecursionlimit(10**6)

from collections import deque #파이썬에서 큐를 사용하기 위한 코드.
#숫자들을 입력한다.
a,b,c = map(int,input().split())

#2차원 배열은 이런 식으로 정의를 한다.
A=[[]for i in range(a+1)]

#bfs,dfs의 결과를 저장할 배열들을 정의한다.

depth=[] #dfs(후입선출)
breadth=[] #bfs(선입선출)
stack=[] #dfs에서 사용할 스택.
q=deque() #bfs에서 사용할 큐.
iscalleddfs=[0]*(a+1)#각 숫자가 호출이 되었는가? 호출시 값을 1로 변환
iscalledbfs=[0]*(a+1)
for i in range(b):
    n,m = map(int,input().split())
    A[n].append(m)
    A[m].append(n)

#print(A) #배열에 입력값이 잘 들어가 있나? 확인

#배열을 정렬한다.
for i in range(a+1):
    A[i]=sorted(A[i])

#print(A)


#dfs를 먼저 정의한다. dfs는 후입선출/스택을 쓴다.
#중요한 것은 처음 다루는 숫자에 한해서 stack에 넣거나 새로운 dfs 함수를 호출한다거나 하는 것이다.
def dfs(n):
    if(len(iscalleddfs)==a):
        return #모두 집어 넣었다면 return으로 호출 끝냄.
    if(iscalleddfs[n]==0): 
        stack.append(n)
        depth.append(n)
        iscalleddfs[n]=1 
    for i in A[n]:
        if(iscalleddfs[i]==0):
            stack.append(i)
            #depth.append(i)
            iscalleddfs[n]=1 
            k=stack.pop()
            dfs(k)


#그 다음에 bfs를 정의한다. bfs는 선입선출/큐를 쓴다.
def bfs(n):
    if(len(iscalledbfs)==a):
        return #모두 집어넣었다면 return으로 호출을 끝냄.
    if(iscalledbfs[n]==0):
        breadth.append(n)
        q.append(n)
        iscalledbfs[n]=1
    for i in A[n]:
        if(iscalledbfs[i]==0):
            q.append(i)
            #breadth.append(i)
            iscalledbfs[n]=1
    q.popleft() #큐의 가장 먼저 들어가 있는 놈을 꺼낸 뒤 지운다.
    if len(q)==0:
        return
    bfs(q[0])#dfs와 재귀함수를 호출하는 타이밍이 다르다.    
    
#값을 출력한다.
dfs(c)
bfs(c)

print(*depth)
print(*breadth)
```

```python
#백준1715번(골드4)-2024/07/17

#아이디어
#1.문제 자체는 굉장히 쉬운 편이다.
#2.하지만, 우선순위 큐의 개념을 써야 해서, 이에 대한 학습이 필요하다.
#3.우선순위 큐를 쓰는 이유는 sort하는 데 많은 시간에 걸리기 때문.

from queue import PriorityQueue #우선순위 큐를 import한다.

N = int(input())


pq=PriorityQueue() #우선순위 큐에 쓸 변수를 정의한다.
s=0 #출력값. sum의 약자이다.


for i in range(N):
    a=int(input())
    pq.put(a) #pq에 a를 추가한다.
    
    
#우선순위큐에서 값을 빼낸다. 카드 묶음이 1개 남을 때까지 반복한다.
while pq.qsize()>1:
    a=pq.get()
    b=pq.get()
    c=a+b
    pq.put(c)#c를 한번 더 put해야 한다는 것을 잊지 말자.
    s=s+c #s에 c을 더함으로서 값을 누적시킨다.
    
    

    
print(s)
```

```python
#백준13023번(골드5)-2024/07/20


#아이디어
#이 문제는 dfs로 푸는 문제이다.
#1.for i in range(n)을 정의한다. visited 배열도 정의한다.
#2.A[i]의 숫자들을 탐방한다. visited가 방문하지 않은 곳으로 간다.
#3.방문시 count++하고 A[k]로 이동한다.
#4.2,3을 반복한다. 
#5.dfs의 count가 5가 될 때 1을 출력한다.
#6.핵심은 dfs(i)가 아니라 dfs(i,count=0)와 같이 정의해서, 깊이 변수를 다루는 것이다. 깊이 변수를 전역 변수로 정의하면 골치아프다.

#오류 및 해결
#1.arrived=1이 되었는데 막상 print을 할 때는 0이 나왔다.arrived를 정수형으로 선언하지 말고 True/False로 선언하면 오류를 막을 수 있다. 또한 변수를 global로 선언해야 한다.
#2.dfs를 선언하고 바로 다음 줄에도 visited[i]=1을 해야 한다. 이걸 까먹으면 섭하지.
#3.dfs에서, 다른 dfs를 호출하지 못했다면 그냥 무위로 돌아간 것이므로, visited를 false로 한다.
#4.백트래킹을 활용해야 한다. 안그러면 다른 경우의 수에서 탐색을 원활히 할 수 없다.

import sys
sys.setrecursionlimit(10000) #재귀 문제를 풀 때 이 변수를 정의하는 것이 절대적이다.
input=sys.stdin.readline


N,M = map(int,input().split())

A=[[] for _ in range(N+1)]

for i in range(M):
    a,b = map(int,input().split())
    A[a].append(b)
    A[b].append(a)


#print(A) #A가 잘 들어가 있나?

arrived=False #깊이가 5가 된 적이 있나?
visited=[0]*(N+1)
def dfs(i,count):
    visited[i]=1
    global arrived
    #print(count)
    if count==5:
        arrived=True #여기서 arrived를 1로 바꿨는데 밑의 arrived는 0이 나온다. 해결:arrived를 True/False로 바꾼다.
        return
        #print(arrived)
    for k in A[i]:
        if(visited[k]==0):
            #visited[k]=1
            dfs(k,count+1)
    visited[i]=0 #다른 경로에서도 재탐색이 가능하도록 하는 코드. 이를 백트레킹이라고 한다.
for i in range(N):
    #방문했는지 안했는지 탐색하는 배열.
   
    dfs(i,1)
    if arrived:
        break
    
if arrived:
    print(1)
else:
    print(0)
```

```python
#백준1033번(골드2)-2024/07/24


#아이디어
#1.이 문제는 dfs와 최소공배수/최대공약수/유클리드 호제법 등의 개념을 융합하는 문제이다.
#2.'개념의 융합'이라는 것을 생각하지 않을 시 상당히 어려운 문제이다. 2시간이 걸려도 못풀수도...
#3.append을 할 때 ()로 값들을 묶어서 넣자. 그러면 한번에 append가 가능하다.

#풀이방법
#1.2차원 배열/값 배열/방문 배열/최소공배수 배열을 정의한다.
#2.숫자들을 입력받는다.
#3.2차원 배열에 값들을 입력받는다. 이 때, 최소공배수 배열에 값을 넣고 최소공배수를 구한다.
#4.비율에 대한 최소공배수를 업데이트한다.(gcd/유클리드 호제법을 사용한다.)
#5.최소공배수*(p/q)의 값을 값 배열에 저장한다. 이 때, dfs 방식으로 탐색하고,탐색 시 방문 배열에 방문 표시를 한다. 모두 탐색 시 종료한다. 
#6.값 배열을 출력한다. 
#7.append하는 부분에서 인덱스가 바뀌면 비율도 바뀐다는 점을 잘 파악해야 한다.
import sys
sys.setrecursionlimit(10 ** 6)

N = int(input())

A=[[]for i in range(N)] #인접 리스트 정의 

isvisited=[0]*int(N) #이 노드를 방문했는가?
C=[0]*int(N)#값 배열
lcm=1 #최소공배수값


#최대공약수 구하는 함수
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)


for i in range(N-1):
    a,b,p,q = map(int,input().split())
    A[a].append((b,p,q))
    A[b].append((a,q,p)) #인접리스트에 값들을 넣는다. 숫자가 바뀌면 비율도 바뀌는것을 숙지해야 한다.
    
    lcm = lcm*int(p*q/gcd(p,q)) #먼저 두 수의 최대공약수를 구한다.(gcd)그런 다음 lcm의 값을 갱신한다.
    #print("lcm의 값:",lcm)
    
#A가 잘 들어갔나?
#print(A)


#초기값을 구한다
#a = int(B[0]*B[1]//gcd(B[0],B[1])) #//은 정수 나눗셈이다.

    
#초기값을 제외한 부분에서 for문을 돌린다. 
#for i in range(2,N-1):
#    count = int(a*B[i]//gcd(a,B[i]))
#    a=count

#count가 최소공배수가 된다.

#print(count)


#최소공배수*p/q의 값을 값 배열에 저장한다. 이 때, dfs 탐색으로 값을 채우며, 다 채웠을 시 종료한다. 

def dfs(i):
    f=0
   
    for k in A[i]:
        #print("k의 값:",k)
        if(sum(isvisited)==N):
            return #모두 다 값을 갱신했으므로 종료.
    
        #print(k[0]) #프린트가 잘 되어 있는지 확인.
        if(isvisited[k[0]]==0):
            isvisited[k[0]]=1
            #print(k[2],k[1])
            C[k[0]] = int((k[2]/k[1])*lcm)
            #print(C[k[0]])
        f=k[0]
    #배열을 다 돈 뒤 다음 dfs 호출
    dfs(f)


#0번을 우선으로 탐색한다.
dfs(0)

print(*C,end=' ')
    
```

```python
#백준 1033번에 대한 모범 풀이-위의 풀이와 무엇이 다른지 잘 파악하자.
N = int(input())
A = [[]for i in range(N)]
visited = [False]*(N)
D = [0]*(N)
lcm = 1


def gcd(a,b):  #최대공약수 함수 구현
    if b==0:
        return a
    else:
        return gcd(b,a%b)
        
        
def DFS(v): #dfs 함수 정의
    visited[v] = True #dfs가 돌면 방문한 곳을 true로 해야 한다.
    for i in A[v]:
        next = i[0] #다음 행선지를 미리 정한다.
        if not visited[next]: 
            #다음 행선지가 방문하지 않은 곳이라면
            D[next] = D[v]*i[2]//i[1] #D에 값을 넣는다.
            DFS(next) #재귀함수 호출

for i in range(N-1):
    a,b,p,q = map(int,input().split()) #숫자들 입력
    A[a].append((b,p,q))
    A[b].append((a,q,p)) #한번에 여러개의 숫자를 집어넣는다.
    lcm = lcm*(p*q//gcd(p,q)) #//은 정수 나누기이다.
    
D[0]=lcm #최소공배수를 배열에 저장
DFS(0) #최소공배수를 D[0]에 저장을 했으므로 dfs(0)로 정의한다.

mgcd=D[0] #계산이 끝나고 인덱스 0의 값을 mgcd에 넣는다. 
            

#모든 수의 최대공약수를 구한다.          
for i in range(1,N):
    mgcd = gcd(mgcd,D[i])

#출력 
for i in range(N):
    #각 노드의 값을 모든 노드의 최대공약수로 나눈 뒤 출력한다.    
    print(int(D[i]//mgcd),end=' ')
    
```
